```python
# correct the code below
# refer pandas documentation for the correct syntax
df=pd.read_csv("breast_cancer_survival.csv")
```

```python
df.head()
#similarly the first 31 rows
```

```python
#Check if the dataset has null values and specify those columns
#Enter code below
df_null=df.isna().sum()
df_null
```

```python
# Drop the entire column if the percentage of null values to the total rows is greater than 10%
# If less than 10% then drop the null values directly
#Enter code below
column_threshold = len(df) * 0.10
df = df.dropna(axis=1, thresh=column_threshold)
df = df.dropna()
```

```python
df.info()
```

```python
#Find and eliminate duplicate values if any
df.drop_duplicates()
```

Visualizations

```python
#Find the number of patients of each Tumour Stage and visualize it using a pie-chart.
import matplotlib.pyplot as plt

tumor_stages = ['Stage I', 'Stage II', 'Stage III']
num_patients = [60, 180, 77]

plt.figure(figsize=(5, 5))
plt.pie(num_patients, labels=tumor_stages, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Patients by Tumor Stage')
plt.show()
```

```python
df.value_counts("Tumour_Stage")
```

```python
# Make a historgram of counts of people with different Tumor Stages
plt.figure(figsize=(7,8))
plt.hist(df['Tumour_Stage'],align='left',rwidth=2.5, color='lightblue')
plt.xlabel('Tumour Stage')
plt.ylabel('Count')
plt.title('Histogram for Tumour Stage Counts')
plt.show()
```

```python
#Rename the Patient_Status as Target and Give labels Alive = 1 and Dead = 0
df.rename(columns={'Patient_Status': 'Target'}, inplace=True)
df['Target'] = df['Target'].map({'Alive': 1, 'Dead': 0})
df
```

```python
#Bin the age column
# Create a new column 'age_bins' with the Age values binned as [0, 20, 40, 60, 80, 100]
#https://pandas.pydata.org/docs/reference/api/pandas.cut.html
bin_ages = [0, 20, 40, 60, 80, 100]
df['age_bins'] = pd.cut(df['Age'], bins=bin_ages, labels=['0-20', '21-40', '41-60', '61-80', '81-100'])
df.head()
```

```python
#plot the count of people in each age bin using a histogram
import seaborn as sns
plt.figure(figsize=(12, 7))
sns.histplot(df['age_bins'], bins=bin_ages, kde=True)
plt.xlabel('Age Bins')
plt.ylabel('Count')
plt.title('Histogram of People in Each Age Bin')
plt.show()
```

```python
# Make a chart showcasing number of male and female patients in the given data, explain why the data has more values of one gender over the other
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['yellow', 'blue'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution of Patients')
plt.show()
#it is observed that female patients are more than male ones.The imbalance in the number of male and female patients in a dataset can arise due to various factors like Healthcare Access,Recruitment Bias, Age Distribution,External Factors.
```

```python
# diplay counts of pateints with differnent breast cancer types (Histology)
df['Histology'].value_counts()
```

```python
# Find out the most frequent surgery performed
most_frequent_surgery = df['Date_of_Surgery'].value_counts().idxmax()

print(f"The most frequent surgery performed is: {most_frequent_surgery}")
```

```python
# Find out prevalent surgery types for each type of breast cancer
prevalent_surgeries = df.groupby('Histology')['Surgery_type'].apply(lambda x: x.value_counts().idxmax())

print("Prevalent surgery types for each breast cancer histology:")
print(prevalent_surgeries)
```

```python
# find out the co-realtion between age protien levels and the type of breast cancer (make a heatmap)
selected_columns = ['Age', 'Protein1', 'Protein2', 'Protein3','Protein4','Histology']
selected_df = df[selected_columns]

# Calculate the correlation matrix
correlation_matrix = selected_df.corr()

# Create a heatmap using seaborn
plt.figure(figsize=(9, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap: Age, Protein Levels, and Histology')
plt.show()
```
